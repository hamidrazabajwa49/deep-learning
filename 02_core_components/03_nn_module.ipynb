# nn.Module — Building Neural Networks
Three patterns for defining models: explicit layers, manual forward, and `nn.Sequential`.

## Setup

```python
import torch
import torch.nn as nn
from torchinfo import summary
```

## Pattern 1 — Single Linear Layer (Logistic Regression)

```python
class LogisticModel(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.linear(x))
```

```python
features = torch.rand(10, 5)
model = LogisticModel(features.shape[1])
print(model(features))
summary(model, input_size=features.shape)
```

## Pattern 2 — Hidden Layer (Explicit)

```python
class TwoLayerModel(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.fc1 = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        return self.sigmoid(self.fc2(x))
```

```python
features = torch.rand(10, 5)
model = TwoLayerModel(features.shape[1])
print(model(features))
summary(model, input_size=features.shape)
```

## Pattern 3 — Sequential

```python
class SequentialModel(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.network(x)
```

```python
features = torch.rand(10, 5)
model = SequentialModel(features.shape[1])
print(model(features))
summary(model, input_size=features.shape)
```
