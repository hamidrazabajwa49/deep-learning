# Autograd
PyTorch automatic differentiation engine. Computes gradients via reverse-mode AD for arbitrary computation graphs.

## Setup

```python
import torch
```

## Scalar Gradients

```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
y.backward()
print(f"dy/dx at x=3: {x.grad}")  # Expected: 6.0
```

```python
# Chain rule: z = sin(x^2), dz/dx = 2x * cos(x^2)
x = torch.tensor(4.0, requires_grad=True)
z = torch.sin(x ** 2)
z.backward()
print(f"dz/dx at x=4: {x.grad:.6f}")
```

## Manual vs Autograd Gradients — Binary Classifier

```python
def binary_cross_entropy(pred, target, eps=1e-8):
    pred = torch.clamp(pred, eps, 1 - eps)
    return -(target * torch.log(pred)) - (1 - target) * torch.log(1 - pred)
```

```python
# Manual gradients
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0)
b = torch.tensor(0.0)

z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)

dL_dy_pred = (y_pred - y) / (y_pred * (1 - y_pred))
dy_pred_dz = y_pred * (1 - y_pred)

dL_dw = dL_dy_pred * dy_pred_dz * x
dL_db = dL_dy_pred * dy_pred_dz * 1

print(f"Manual  dL/dw: {dL_dw:.6f}")
print(f"Manual  dL/db: {dL_db:.6f}")
```

```python
# Autograd gradients
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)
loss.backward()

print(f"Autograd dL/dw: {w.grad:.6f}")
print(f"Autograd dL/db: {b.grad:.6f}")
```

## Vector Gradients

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = (x ** 2).mean()
y.backward()
print(f"d(mean(x^2))/dx = {x.grad}")  # Expected: [2/3, 4/3, 2]
```
